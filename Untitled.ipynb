```python

def converter(x):
    list1=[]
    for i in x:
        if(i[0]>=0.6):
            list1.append('Highly_Fatal_And_Damaging')
        elif(i[0]+i[1]>=0.6):
            list1.append('Minor_Damage_And_Injuries')
        elif(i[0]+i[1]+i[2]>=0.6):
            list1.append('Significant_Damage_And_Fatalities')
        else:
            list1.append('Significant_Damage_And_Serious_Injuries')
    return list1

```
